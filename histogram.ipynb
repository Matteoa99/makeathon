```python
import matplotlib.pyplot as plt
```

```python
# import urllib library
from urllib.request import urlopen
  
# import json
import json
# store the URL in url as 
# parameter for urlopen
url = "https://raw.githubusercontent.com/Matteoa99/makeathon/main/data/unified_review.json"
  
# store the response of URL
response = urlopen(url)
  
# storing the JSON response 
# from url in data
data_json = json.loads(response.read())
```

```python
data_json[0]
```

```python
import numpy as np

plt.figure(figsize=(8,4))

plt.hist([a['rating'] for a in data_json], bins=np.arange(0, 7, 1), color="#AA0000")

plt.xlabel("Score")
plt.ylabel("Counts")

plt.xticks(np.arange(1.5, 6.5, 1), labels=range(1, 6))
plt.xlim(1, 6)

plt.show()
```

```python
random_scores = np.random.rand(len(data_json))
for i in range(len(data_json)):
    data_json[i]['random_score'] = random_scores[i]
```

```python
import numpy as np
import matplotlib as mpl
from matplotlib import pyplot as plt

category_names = ['Positive', 'Neutral', 'Negative']
results = {
    'Topic 1': [30, 10, 60],
    'Topic 2': [20, 20, 60],
    'Topic 3': [45, 5, 50],
    'Topic 4': [60, 10, 30],
    'Topic 5': [45, 15, 40],
    'Topic 6': [80, 5, 15]
}


def survey(results, category_names):
    """
    Parameters
    ----------
    results : dict
        A mapping from question labels to a list of answers per category.
        It is assumed all lists contain the same number of entries and that
        it matches the length of *category_names*.
    category_names : list of str
        The category labels.
    """
    labels = list(results.keys())
    data = np.array(list(results.values()))
    data_cum = data.cumsum(axis=1)
    category_colors = ["#94b832", "#3f9cb1", "#be1c24"]
    #(
    #    np.linspace(0.15, 0.85, data.shape[1]))

    fig, ax = plt.subplots(figsize=(9.2, 5))
    ax.invert_yaxis()
    ax.xaxis.set_visible(False)
    ax.set_xlim(0, np.sum(data, axis=1).max())

    for i, (colname, color) in enumerate(zip(category_names, category_colors)):
        widths = data[:, i]
        starts = data_cum[:, i] - widths
        rects = ax.barh(labels, widths, left=starts, height=0.5,
                        label=colname, color=color)

        #r, g, b, _= color
        #text_color = 'white' if r * g * b < 0.5 else 'darkgrey'
        text_color = 'white'
        #ax.bar_label(rects, label_type='center', color=text_color)
    ax.legend(ncol=len(category_names), bbox_to_anchor=(0, 1),
              loc='lower left', fontsize='small')

    return fig, ax


survey(results, category_names)
plt.show()
```
